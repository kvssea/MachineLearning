```python
#GRADIENT DESCENT
#The idea is to minimize a function J(theta0, theta1) by changing theta 0 and theta1 to try and reduce J(theta0, theta1) until we reach a local minumum.
#The formula for gradient descent is as follows:
```

$$ \huge \theta_{j} := \theta_{j} - \alpha \frac{\partial}{\partial \theta_{j}}J(\theta_{0}, \theta_{1}) \quad \quad \quad \quad (\textit{for j = 0 and j = 1})$$

```python
#Gradient descent takes an iterative approach of taking a step downwards to find the local minimum.  It relies on the slope of the curve (the derivative) in that direction with the steepest descent.
#Alpha is the learning rate and determine how large of a step we take downwards in the process there are two values for j (0 and 1) because we have one feature and one base number (the y intercept theta0) as this is only a single variable linear regression formulat we're dealing with, however this generalizes to additional features as in multivariable linear regression
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
```

```python
def hyp_fxn(x, theta0 = 0, theta1 = 0):
    return(theta0 + (theta1 * x))
```

```python
def cost_fxn(xlist, ylist, theta0 = 0, theta1 = 0):
    sums = 0
    for i in range(len(xlist)):
        sums += (hyp_fxn(x = xlist[i], theta0 = theta0, theta1 = theta1) - ylist[i])**2
    sums = sums / (2 * len(xlist))
    return(sums)
```

```python
x = [1, 2, 3] #set sample values of x and y in which we know the cost function value of 0 should be met at theta0 = 0, theta1 = 1 since it is a 100% correlation / fit to the line at these meaures
y = [1, 2, 3]
cost_fxn(x,y, theta0 = 0, theta1 = 1) #call cost function to ensure we are getting the expected 0.0 value
```

```python
theta0 = list(range(-10,10)) #set a series of theta0s to test
theta1 = list(range(-10,10)) #set a series of theta1s to test
J = [] #create an output list of cost function theta0, theta1 inputs
for thetazero in theta0:
    for thetaone in theta1:
        J.append(cost_fxn(xlist = x, ylist = y, theta0 = thetazero, theta1 = thetaone))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = [15,10])
ax = fig.add_subplot(111, projection='3d')

thetax, thetay = np.meshgrid(theta0, theta1)
thetaz = np.reshape(J, (len(theta0),len(theta1)))
ax.set_xlabel('$\u03B8_0$', fontsize = 14)
ax.set_ylabel('$\u03B8_1$', fontsize = 14)
ax.set_zlabel('J($\u03B8_0$, $\u03B8_1$)', fontsize = 14)
ax.set_title('Plot of Cost Function for Given \u03B8s')
ax.plot_surface(thetax, thetay, thetaz, cmap = 'viridis')
print(J[0:10]) #print to verify correct J values match the plotted Z values
print(thetaz[0]) #print to verify correct J values match the plotted Z values
cost = cost_fxn(xlist = x, ylist = y, theta0 = 0, theta1 = 1)
print(f'Amount of error at the optimized value = {cost:.2f}')
```

```python
#Use gradient descent to determine optimized value (minimum) of J(theta0, theta1)
```

$$ \huge \theta_{j} := \theta_{j} - \alpha \frac{\partial}{\partial \theta_{j}}J(\theta_{0}, \theta_{1}) \quad \quad \quad \quad (\textit{for j = 0 and j = 1})$$

```python
#Deriving the gradient descent algorithm
# 
# Recall the cost functions for J(theta0, theta1)
```

$$ \huge J(\theta_{0}, \theta_{1}) = \frac{1}{2m}\huge \sum (h_{0}(x^{i}) - y^{i})^{2}$$

```python
#Subsitute theta0, theta1 values for the hypothesis function to put it terms of thetas
```

$$ \huge J(\theta_{0}, \theta_{1}) \frac{1}{2m} \sum \limits_{i=1} ^{m} (\theta_{0} + \theta_{1}x^{(i)}) - y^{(i)})^{2}$$

```python
#Take the partial derivative of the cost function with respect to theta1
```

$$ \huge\frac{\partial}{\partial \theta_{1}} J(\theta_{0}, \theta_{1}) = \theta_{1} := \frac{1}{m} \sum \limits_{i=1} ^{m} (\theta_{0} + \theta_{1}x^{(i)} - y^{(i)})\cdot x^{(i)}$$

```python
#Take the partial derivative of the cost function with respect to theta0
```

$$ \huge\frac{\partial}{\partial \theta_{0}} J(\theta_{0}, \theta_{1}) = \theta_{0} := \frac{1}{m} \sum \limits_{i=1} ^{m} (\theta_{0} + \theta_{1}x^{(i)} - y^{(i)})$$

```python
#Repeat the algorithms for theta0 and theta1 below until convergence
```

$$ \huge \theta_{1} := \theta_{1} - \alpha \frac{1}{m} \sum \limits_{i=1} ^{m} ((\theta_{0} + \theta_{1}x^{(i)}) - y^{(i)})\cdot x^{(i)}$$

$$ \huge \theta_{0} := \theta_{0} - \alpha \frac{1}{m} \sum \limits_{i=1} ^{m} ((\theta_{0} + \theta_{1}x^{(i)}) - y^{(i)})$$

```python
#Let's choose a value for theta0, theta1 to apply the algorithm to: 
#I'll choose theta0 = 3, theta1 = -3 for the sample simple sets of x and y used previously

x = np.array([1,2,3])
y = np.array([1,2,3])


def gradient_descent(x, y, a, i, theta0 = 0, theta1 = 0):
    plt.rcParams['figure.figsize'] = [14, 6] 
    fig, (ax1,ax2,ax3) = plt.subplots(1,3)
    
    m = len(x)
    current_theta0 = theta0
    current_theta1 = theta1
    plot_alpha = i
    for i in range(i):
        current_theta0 = current_theta0 - (a / m) * sum((current_theta0 + (current_theta1*x)) - y)
        current_theta1 = current_theta1 - (a / m) * sum((current_theta0 + (current_theta1*x) - y) * x)
        cost = 1 / (2*m) * sum((current_theta0 + (current_theta1 * x) - y)**2)

        ax1.plot(i,cost, '.b',color = '#00DE6D')
        ax1.set_xlabel('# of iterations (i)')
        ax1.set_ylabel('J(\u03B80, \u03B81) (Cost)')


        ax2.plot(i, current_theta0, '.b',color = '#00DE6D')
        ax2.set_xlabel('# of iterations (i)')
        ax2.set_ylabel('\u03B80')


        ax3.plot(i,cost, '.b',color = '#00DE6D')
        ax3.set_xlabel('# of iterations (i)')
        ax3.set_ylabel('\u03B81') 

    print(f'\u03B80 = {current_theta0:.4f}, \u03B81 = {current_theta1:.4f}, cost = {cost:.4f}')
        
    
gradient_descent(x, y, a = 0.05, i = 2000, theta0 = 3, theta1 = -3)
#Here we see the gradient descent indicates the cost function reducing, along with the values of theta0, theta1 at each iteration.  
```
